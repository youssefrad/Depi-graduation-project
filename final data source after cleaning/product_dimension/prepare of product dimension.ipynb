# here modify missing  colum product_id 

```python
import pandas as pd
import numpy as np


# Step 1: Initialize a counter
next_id = 1


# Step 2: Iterate through the DataFrame and fill NaN values
for index in df.index:  # Use df.index to get the actual index labels
    if pd.isna(df.loc[index, 'product_id']): # as df.loc[index, 'product_id'].isna()
        # Replace with the current next_id and increment
        df.loc[index, 'product_id'] = next_id
        next_id += 1  # Increment next_id only when filling NaN
    else:
        # If the current product_id is not NaN, increment the next_id to the next integer
        if df.loc[index, 'product_id'] >= next_id:
            next_id = int(df.loc[index, 'product_id']) + 1 # because here is float 

# Step 3: Convert product_id back to integer type if needed
df['product_id'] = df['product_id'].astype(int)

print("\nUpdated DataFrame with missing product_ids filled:")
df
```

# here modify the category colum 

```python
import numpy as np
df['category'].unique() # to determine actual data at category colum 

choices=['home goods','food','clothing','electronics','beauty']

#check for nulls values 
unlls_nums=df['category'].isna().sum()
print(f"before cleaning: {unlls_nums}")

#clean data
unlls_num=df['category'].isna()
df.loc[unlls_num,'category']=np.random.choice(choices,size=unlls_nums)

#check for cleaning 
unlls_nums=df['category'].isna().sum()
print(f"after cleaning: {unlls_nums}")
```

```python
# chech all things is done 
print(df.isna().sum())
print(df['category'].unique())
df[['category']]
```

# here modify dimensions colum

```python
import numpy as np
df['dimensions'].unique() # to determine actual data at category colum 

choices=['medium','large','small']

#check for nulls values 
unlls_nums=df['dimensions'].isna().sum()
print(f"before cleaning: {unlls_nums}")

#clean data
unlls_num=df['dimensions'].isna()
df.loc[unlls_num,'dimensions']=np.random.choice(choices,size=unlls_nums)

#check for cleaning 
unlls_nums=df['dimensions'].isna().sum()
print(f"after cleaning: {unlls_nums}")
```

```python
# chech all things is done 
print(df.isna().sum())
print(df['dimensions'].unique())
df[['dimensions']]
```
