```python
hire_date = sales_person.pop('hire_date')

sales_person.insert(4 ,'hire_date' ,hire_date)

sales_person
```

```python
# delete hire colu because it have so issues 
sales_person.drop(columns=['hire_date'],inplace=True)

sales_person
```

```python
import numpy as np

np.random.seed(0)

sales_person['commission_rate'] = np.random.uniform(0, 20, size=len(sales_person))
sales_person['commission_rate'] = sales_person['commission_rate'].astype(int)
sales_person
```

```python
sales_person = sales_person.rename(columns = {'sales_target': 'total_sales'})
```

```python
sales_person['commission_earned'] = sales_person['commission_rate'] * sales_person['total_sales'] / 100
sales_person['target_met'] = sales_person['total_sales'] >= 120000
```

```python
sales_person
```

```python
commission_earrned = sales_person.pop('commission_earned')

sales_person.insert(11, 'commission_earned', commission_earrned)

target_met = sales_person.pop('target_met')

sales_person.insert(12, 'target_met', target_met)

sales_person
```

```python
sales_person.to_json('sales person dimension.json', orient='records', indent=4)
```

```python
# Check for duplicate sales_person_id
duplicates = sales_person[sales_person.duplicated(subset=['sales_person_id'], keep=False)]

# Display the duplicate rows
if not duplicates.empty:
    print("Duplicate sales_person_id found:")
    print(duplicates)
else:
    print("No duplicate sales_person_id found.")
```
