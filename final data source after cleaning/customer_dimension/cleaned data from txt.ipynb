# clean  membership_status  attribute 

```python
import numpy as np
# to determine the unique value
df['membership_status'].unique() 

#replacr and clean data 
replace=['active','pending','inactive']

uum_nulls=df['membership_status'].isna().sum()
print(f'number of nulls before cleaning: {uum_nulls}')

num_null=df['membership_status'].isna()
df.loc[num_null,'membership_status']=np.random.choice(replace,size=uum_nulls)


uum_nulls=df['membership_status'].isna().sum()
print(f'number of nulls after cleaning : {uum_nulls}')


df.isna().sum()
```

# clean join_date attribute 

```python

# to get this method random.choice
import numpy as np

#to show unique value 
df['join_date'].unique() # only one 

# Choices to replace NaN values
choices = ["11/19/2023", "8/17/2020", "5/3/2020","28/9/2024"]

# Count the number of null values in the branch_name column
num_nulls = df['join_date'].isna().sum()
print("Number of nulls before:", num_nulls)

# Create a mask for NaN values
nan_mask = df['join_date'].isna()

# Fill NaN values in branch_name with random choices
df.loc[nan_mask, 'join_date'] = np.random.choice(choices, size=num_nulls)

# Verify the replacement
num_nulls = df['join_date'].isna().sum()
print("Number of nulls after:", num_nulls)

# Display the updated DataFrame
df

```

```python
# Splitting into Day, Month, and Year
# Split the join_date column into day, month, and year
df[['month', 'day', 'year']] = df['join_date'].str.split('/', expand=True)

# Display the dataframe with day, month, and year
print(df[['join_date', 'day', 'month', 'year']])
df 
```
