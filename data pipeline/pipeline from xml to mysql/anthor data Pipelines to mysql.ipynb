# Transfor

```python
df.dtypes
```

```python
# 1. Convert `release_date` and `expiration_date` to datetime format
df['release_date']=pd.to_datetime(df['release_date'])
df['expiration_date']=pd.to_datetime(df['expiration_date'])


# 2 Convert to pandas string type (only if you're using Pandas >= 2.0)
# Convert specified columns to string type

df['product_name'] = df['product_name'].astype("string")
df['brand'] = df['brand'].astype("string")
df['category'] = df['category'].astype("string")
df['dimensions'] = df['dimensions'].astype("string")

# Verify the changes
print(df.dtypes)
df

```

```python
# 3 Calculate product age (time between release and expiration)
df['product_age(days)'] = (df['expiration_date'] - df['release_date']).dt.days

# Extract the year and find the min and max year
min_year = df['expiration_date'].dt.year.min()
max_year = df['expiration_date'].dt.year.max()
print(f"Minimum year: {min_year}")
print(f"Maximum year: {max_year}")


df
```

```python
# 5. Standardize the `dimensions` (e.g., small, medium, large into numerical values)
dimension_mapping = {'small': 1, 'medium': 2, 'large': 3}
df['dimension_size'] = df['dimensions'].map(dimension_mapping)

df
```
