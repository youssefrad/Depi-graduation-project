# remove duplicate and misasing value 

```python
df_clean1 = df.dropna()
```

```python
df_clean1 = df.drop_duplicates()
```

```python
df_clean1.info()
```

```python
# Convert 'custimer_id' column to int 
df_clean1['CustomerID']=df_clean1['CustomerID'].astype(int)
```

# some modification 

```python
df_clean1.rename(columns={'Country': 'branch_name', 'StockCode': 'product_name'}, inplace=True)
```

```python
df_clean1.info()
```

```python
df_clean1.head()
```

```python
# Generate branch_id based on 'branch_name'

df_clean1['branch_id'] = df_clean1.groupby('branch_name').ngroup() + 1

# Generate product_id based on 'product_name'

df_clean1['product_id'] = df_clean1.groupby('product_name').ngroup() + 1
```

```python
df_clean1['totalsales']= df_clean1['Quantity'] * df_clean1['UnitPrice']
df_clean1['quantity_sold']=df_clean1['Quantity']
```

```python
# Dropping a single column
df_clean1.drop(columns='Quantity', inplace=True)
```

```python
import numpy as np
df_clean1['salesperson_id'] = np.random.randint(1, 100, df_clean1.shape[0])
```

```python
df_clean1.info()
```

```python
# Define the desired column order
desired_order = [
    'InvoiceNo',
    'InvoiceDate',
    'CustomerID',
    'branch_name',
    'branch_id',
    'salesperson_id',
    'product_id',
    'product_name',
    'Description',
    'quantity_sold',
    'UnitPrice',
    'totalsales'  
]
df_clean1=df_clean1[desired_order]
df_clean1.info()
```

```python
df_clean1
```

```python
df_clean1['branch_name'] = df_clean1['branch_name'].replace('Israel', 'Egypt')
```

```python
df_clean1['branch_name'].unique()
```

```python
df_clean1['InvoiceDate'] = pd.to_datetime(df_clean1['InvoiceDate'], unit='ms').dt.date
```
